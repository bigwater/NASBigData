```python
import pandas as pd
```

```python
exps_csv = {
    "AgEBO-LR-BC-#R-Tim-1R": "training_agebov3_8_timeout-1R.csv",
    "AgEBO-LR-BC-#R-Tim-8R": "training_agebov3_8_timeout-8R.csv",
    "AgEBO-LR-BC-#R-Epo-1R": "training_agebov3_8-1R.csv",
    "AgEBO-LR-BC-#R-Epo-8R": "training_agebov3_8-8R.csv"
}
```

```python
line_style = {
    "AgEBO-LR-BC-#R-Tim-1R": ":",
    "AgEBO-LR-BC-#R-Tim-8R": "-",
    "AgEBO-LR-BC-#R-Epo-1R": "-",
    "AgEBO-LR-BC-#R-Epo-8R": ":"
}

color_style = {
    "AgEBO-LR-BC-#R-Tim-1R": "b",
    "AgEBO-LR-BC-#R-Tim-8R": "b",
    "AgEBO-LR-BC-#R-Epo-1R": "r",
    "AgEBO-LR-BC-#R-Epo-8R": "r"
}
```

```python
dfs = {k:pd.read_csv(v) for k,v in exps_csv.items()}
```

```python
plt.figure()

for k,df in dfs.items():
    plt.plot(df.val_acc, label=k, linestyle=line_style[k], color=color_style[k])

plt.xlabel("Epochs")
plt.ylabel("Validation Accuracy")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure()

for k,df in dfs.items():
    T = 0
    times = []
    for d in df.duration:
        T += d
        times.append(T)
    plt.plot(times, df.val_acc, label=k, linestyle=line_style[k], color=color_style[k])

plt.xlabel("Time (sec.)")
plt.ylabel("Validation Accuracy")
plt.legend()
plt.tight_layout()
plt.show()
```
