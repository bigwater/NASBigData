```python
import openml
import pandas as pd
from openml.datasets.functions import get_dataset
```

```python
openml_df = openml.datasets.list_datasets(output_format="dataframe")
openml_df.head(n=10)
```

```python
openml_df[openml_df.name == "airlines"]
```

```python
subdf = openml_df[openml_df.NumberOfInstances > 200000]
subdf = subdf[subdf.NumberOfClasses >= 5]
subdf = subdf[subdf.NumberOfInstancesWithMissingValues == 0]
subdf = subdf[subdf.NumberOfFeatures == subdf.NumberOfNumericFeatures + subdf.NumberOfSymbolicFeatures]
# subdf = subdf[subdf.name == "dionis"]
# subdf = subdf[not("BGN" in subdf.name)]
subdf = subdf[~subdf['name'].str.contains("BNG")]
subdf.sort_values(["NumberOfInstances"], ascending=False).head(n=50)
```
